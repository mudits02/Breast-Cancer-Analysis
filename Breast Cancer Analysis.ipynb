## Converting String Data to Numeric Output Data

```python
# Handling String in Data
#String data are generally not preferred as operations on them or any kind of mathematical operations get hindered by them.
#In the case of Classification , we generally prefer numbering the outputs if they are in string data type
#In this case M = Malignant Tumour and B = Belingn Tumour. M tumours are cancerous while the B tumours are not cancerous
def func(s):
    if s == 'M':
        return 1
    else :
        return 0

df["cancer"] = df.diagnosis.apply(func)
df.head()
print(df["cancer"])
df.iloc[98:100]
print(df[df.diagnosis == 'B'])
df.head()
```

```python
X = df.iloc[ :  , 2 : 32].values
y = df["cancer"].values
print(X)
print(y)
```

## Splitting Test and Training set

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.25 , random_state = 42)
print(y_train)
# '0' represents benign while '1' represents malignant
```

## MODEL TRAINING (SCIKIT LEARNING)

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver = 'liblinear')
model.fit(X_train , y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accu = accuracy_score(y_test , y_pred)
accuracy = round(accu , 5) * 100
print("The accuracy of the model is :",accuracy , "%")
```
